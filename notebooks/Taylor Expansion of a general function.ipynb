```python
from scripts.deriviative_higher_function import taylor
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return 2 * np.sin(x) ** 2 + x
```

```python
nmax = 20
```

```python
h = 0.05
```

```python
plt.ylim([-5,8])
x_list = np.linspace(-5, 5, 101)
plt.scatter(x_list, function(x_list))

plt.plot(x_list, taylor(function, x_list, 0, nmax, h), "blue")
plt.plot(x_list, taylor(function, x_list, 2, nmax, h), "red")
plt.plot(x_list, taylor(function, x_list, -3, nmax, h), "green")
```

```python
def function(x):
    return 2*np.sin(x)**2 * np.exp(-0.1*x**2)
```

```python
plt.ylim([-2,3])
x_list = np.linspace(-5, 5, 101)
plt.scatter(x_list, function(x_list))

plt.plot(x_list, taylor(function, x_list, 0, nmax, h), "blue")
plt.plot(x_list, taylor(function, x_list, 2, nmax, h), "red")
plt.plot(x_list, taylor(function, x_list, -3, nmax, h), "green")
```
